```python
df_cleaned=df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
df_cleaned
```

```python
df_cleaned.isnull().sum()
```

```python
#count the number of survived with respect to survived category
```

```python
survived_count=df_cleaned.groupby("Survived")["Survived"].count()
```

```python
plt.figure(figsize=(4,5))
plt.bar(survived_count.index,survived_count.values)
plt.title("Grouped By Survival")
plt.xticks([0,1],["Not Survived","Survived"])
```

```python
p_class_count=df_cleaned.groupby("Pclass")["Pclass"].sum()
p_class_count
```

```python
plt.figure(figsize=(7,7))
plt.pie(p_class_count.values,labels=['Class A','Class B','Class C'])
plt.title("Grouped By P-Class")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(p_class_count.values,labels=['Class A','Class B','Class C'],autopct='%1.1f%%',textprops={'fontsize':13})
plt.title("Grouped By P-Class")
plt.show()
```

```python
df_gender=df_cleaned.groupby("Gender")["Gender"].count()
df_gender
```

```python
plt.figure(figsize=(4,4))
plt.pie(df_gender.values,labels=["Male","Female"],autopct='%1.1f%%',textprops={'fontsize':13})
plt.title("Grouped by gender")
plt.show()
```

```python
df_embarked=df_cleaned.groupby("Embarked")["Embarked"].count()
df_embarked
```

```python
plt.figure(figsize=(4,4))
plt.pie(df_embarked.values,labels=["C","Q","S"],autopct='%1.1f%%',textprops={'fontsize':13})
plt.title("Grouped by embarked")
plt.show()
```
