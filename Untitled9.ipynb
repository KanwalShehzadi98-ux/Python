```python
country_count=df.groupby("country")["country"].count()
country_count
```

```python
import seaborn as sns
plt.figure(figsize=(12,9))
plt.plot(country_count.index,country_count.values)
plt.title("Grouped By Country")
plt.xticks(rotation=45,ha='right')
```
