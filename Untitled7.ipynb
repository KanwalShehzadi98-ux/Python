```python
df["Last_Name"]=df["Last_Name"].str.strip("_")
df["Last_Name"]=df["Last_Name"].str.strip("...")
df["Last_Name"]=df["Last_Name"].str.strip("/")
df["Last_Name"]
```

```python
df["Phone_Number"]
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace("|","-")
df["Phone_Number"]=df["Phone_Number"].str.replace("/","-")
df["Phone_Number"]=df["Phone_Number"].str.replace("N-a","0")
df["Phone_Number"]=df["Phone_Number"].fillna(0)
df["Phone_Number"]
```

```python
import pandas as pd

def my_phone_number(num):
    if pd.isna(num):
        return num
    my_num = str(num)
    if len(my_num) >= 6 and '-' not in my_num:
        return f"{my_num[:3]}-{my_num[3:6]}-{my_num[6:]}"
    return my_num

df["Phone_Number"] = df["Phone_Number"].apply(my_phone_number)
df["Phone_Number"]
```
