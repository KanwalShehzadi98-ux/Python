<h3>Title and Axes Labels</h3>
xlabel<br>ylabel<br>title

```python
months = ['Jan','Feb','Mar','Apr','May','June','July']
sales = [10789,12897,13554,12650,14320,15236,234523]

plt.plot(months,sales)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
```

```python
salesA = [10789,12897,13554,13650,14320,15236]
salesB = [10346,13326,14569,14987,15005,15241]
salesC = [9891,11369,11659,12458,12430,13985]
months = ['Jan','Feb','Mar','Apr','May','June']

plt.plot(months,salesA)
plt.plot(months,salesB)
plt.plot(months,salesC)

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison')
plt.legend(labels=['Company A','Company B','Company C'],loc='upper left')
```
